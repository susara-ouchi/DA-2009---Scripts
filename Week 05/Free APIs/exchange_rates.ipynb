### ExchangeRates API

```python
import pandas as pd

api_url = "https://api.exchangerate-api.com/v4/latest/USD"

df_json = pd.read_json(api_url)
df_json.to_csv("Exports/ExchangeRates.csv")
```

### CatNinja API

```python
import requests

api_url = "https://catfact.ninja/fact"
response = requests.get(api_url)
data = response.json()
data

df = pd.DataFrame([data])
df.to_csv("Exports/CatNinja.csv",index=False)
```
