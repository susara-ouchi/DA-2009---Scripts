```python
from bs4 import BeautifulSoup
import pandas as pd
```

Importing file

```python
file_path = "../005 - Portfolio/index.html"

with open(file_path,'r',encoding='utf-8') as file:
    html = file.read()
```

Create the soup

```python
soup = BeautifulSoup(html,'html.parser')
print(soup.prettify())
```

# Extract nav items

```python
nav_items = soup.select("nav .nav-links li a")
nav_items
```

```python
for i in nav_items:
    dict_items = {"Section":i.text.strip(), "Link":i.get("href")}
    print(dict_items)
```

```python
nav_items = soup.select("nav .nav-links li a")
nav_data = [{"Section":link.text.strip(), "Link":link.get("href")} for link in nav_items]
nav_data
```

```python
nav_df = pd.DataFrame(nav_data)
nav_df
```
