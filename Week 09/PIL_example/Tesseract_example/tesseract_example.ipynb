```python
# Install Pytesseract if not installed
# !pip install pytesseract pillow

from PIL import Image
import pytesseract
from IPython.display import display

# 1. Load image (can be preprocessed)
img = Image.open("processed_example.jpg")
display(img)  # Show the image inline

# 2. Basic text extraction
text = pytesseract.image_to_string(img)
print("Extracted Text:\n")
print(text)

# 3. Extract text from a specific region (crop image first)
cropped_img = img.crop((50, 50, 300, 200))  # left, top, right, bottom
region_text = pytesseract.image_to_string(cropped_img)
print("\nText from cropped region:\n")
print(region_text)

# 4. Extract only digits
digits = pytesseract.image_to_string(img, config='--psm 6 -c tessedit_char_whitelist=0123456789')
print("\nDigits only:\n")
print(digits)

# 5. Get detailed data (like bounding boxes)
data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
print("\nOCR Data keys:", data.keys())  # Keys: level, page_num, block_num, par_num, line_num, word_num, left, top, width, height, conf, text
```
