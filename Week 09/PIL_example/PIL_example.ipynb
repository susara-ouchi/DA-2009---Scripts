```python
from PIL import Image, ImageFilter, ImageDraw, ImageFont
from IPython.display import display  # To display images inline in Jupyter

# 1. Load an image
img = Image.open("dog.jpg")
img.show()  # Display the image externally
```

```python
display(img) # Display image in Jupyter notebook itself
```

```python
# 2. Resize the image
resized_img = img.resize((300, 300))
display(resized_img)
```

```python
# 3. Crop the image
cropped_img = img.crop((100, 70, 130, 95))  # (left, top, right, bottom)
display(cropped_img)
```

```python
# 4. Rotate the image
rotated_img = img.rotate(45)  # Rotate 45 degrees
display(rotated_img)
```

```python
# 5. Convert to grayscale
gray_img = img.convert("L")
display(gray_img)
```

```python
# 6. Apply a filter (e.g., blur)
blurred_img = img.filter(ImageFilter.BLUR)
display(blurred_img)
```

```python
# 7. Draw text on image
img_copy = img.copy()
draw = ImageDraw.Draw(img_copy)
font = ImageFont.load_default()
draw.text((10, 10), "Hello PIL! This is a DOG", fill="red", font=font)
display(img_copy)
```

```python
# 8. Save the modified image
img_copy.save("modified_example.jpg")
```
